```python
import tensorflow as tf
import numpy as np
import pandas as pd

mnist = tf.keras.datasets.mnist
tf.reset_default_graph()
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
print(x_train.shape)
x_train=x_train.reshape(x_train.shape[0],28*28)
x_test=x_test.reshape(x_test.shape[0],28*28)
y_train=pd.get_dummies(y_train)
y_test=pd.get_dummies(y_test)

print(y_train.shape)
## Define a Logistic Regression in TF
### y = sigmoid(Wx+b)
## Dim(b) = 10
## Dim(x) = 784
## Dim(Wx) = 10

```

```python
x=tf.placeholder(tf.float32,(None,784))
y=tf.placeholder(tf.float32,(None,10))
pred_x=tf.placeholder(tf.float32,(None,784))

W=tf.Variable(tf.random_normal((784,10)))
b=tf.Variable(tf.random_normal((10,)))

y_logit = tf.squeeze(tf.matmul(x, W) + b)

entropy = tf.nn.sigmoid_cross_entropy_with_logits(logits=y_logit, labels=y)
l = tf.reduce_sum(entropy)

train_op = tf.train.AdamOptimizer(.01).minimize(l)

n_steps = 100

pred_y = tf.nn.softmax(tf.matmul(pred_x,W) + b)
pred=tf.argmax(pred_y, 1)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer()) # Train model
    for i in range(n_steps):
        feed_dict = {x: x_train, y: y_train}
        _, loss = sess.run([train_op, l], feed_dict=feed_dict) 
        print("Current Loss: %f" %loss)
    writer = tf.summary.FileWriter('./graph',sess.graph)
    preds=sess.run(pred,feed_dict={pred_x:x_test})
print(preds
```

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import tensorflow as tf
import numpy as np
import pandas as pd

mnist = tf.keras.datasets.mnist
tf.reset_default_graph()
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
x_train=x_train.reshape(x_train.shape[0],28*28)
x_test=x_test.reshape(x_test.shape[0],28*28)
y_train=pd.get_dummies(y_train)
y_test=pd.get_dummies(y_test)

### Y=Act(W*Act(W*Act(...+b)+b)+b)
### Dim(y)*10
### Dim(Act(W*Act(...)+b)+b)*10

### Dim(Act(W*x_input+b))=N
### Dim(b) = N
### Dim(x_input) = 784
### Dim(W)

x=tf.placeholder(tf.float32,(None,784))
y=tf.placeholder(tf.float32,(None,10))
x_pred=tf.placeholder(tf.float32,(None,784))

w1=tf.Variable(tf.random_normal((784,1024)))
b1=tf.Variable(tf.random_normal((1024,)))

w2=tf.Variable(tf.random_normal((1024,512)))
b2=tf.Variable(tf.random_normal((512,)))

w3=tf.Variable(tf.random_normal((512,64)))
b3=tf.Variable(tf.random_normal((64,)))

w4=tf.Variable(tf.random_normal((64,10)))
b4=tf.Variable(tf.random_normal((10,)))

layer1=tf.matmul(x,w1)+b1
layer1_activation=tf.nn.relu(layer1)

layer2=tf.matmul(layer1_activation,w2)+b2
layer2_activation=tf.nn.relu(layer2)

layer3=tf.matmul(layer2_activation,w3)+b3
layer3_activation=tf.nn.relu(layer3)

layer4=tf.matmul(layer3_activation,w4)+b4
layer4_activation=tf.nn.softmax(layer4)

entropy = tf.nn.sigmoid_cross_entropy_with_logits(logits=layer4, labels=y)
l = tf.reduce_sum(entropy)
train_op = tf.train.AdamOptimizer(.01).minimize(l)
n_steps = 13
#pred_y = tf.nn.softmax(tf.matmul(pred_x,W) + b)
pred=tf.argmax(layer4_activation, 1)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer()) # Train model
    for i in range(n_steps):
        feed_dict = {x: x_train, y: y_train}
        _, loss = sess.run([train_op, l], feed_dict=feed_dict)
        print("Current Loss: %f" %loss)
    writer = tf.summary.FileWriter('./graph', sess.graph)
    preds=sess.run(pred,feed_dict={x:x_test})
print(preds)
```

```python
from keras.models import Sequential
from keras.layers import Activation
from keras.optimizers import Adam
from keras.layers import Dense

model = Sequential()
model.add(Dense(1024, input_dim=784,init='normal',activation='relu'))
model.add(Dense(512, activation = 'relu',kernel_initializer = 'normal'))
model.add(Dense(64,activation = 'relu', kernel_initializer = 'uniform'))
model.add(Dense(10))
model.add(Activation('softmax'))
optimizer = Adam(lr=0.01)
model.compile(loss='binary_crossentropy',optimizer = optimizer, metrics = ['accuracy'])
model.fit(x_train,y_train,epochs = 13, batch_size=128,verbose=1)
```

```python
import time
import sklearn.preprocessing
import os
import math


import tensorflow as tf
import numpy as np
import pandas as pd

mnist = tf.keras.datasets.mnist
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

label_binarizer = sklearn.preprocessing.LabelBinarizer()
y_train = label_binarizer.fit_transform(y_train)
print(x_train.shape)
batch_size=128
learning_rate=0.002

#sess=tf.Session()
X=tf.placeholder(tf.float32,[None,28,28])
Y=tf.placeholder(tf.float32,[None,10])
X_img=tf.reshape(X,[-1,28,28,1])

## Convolutional w. dropout Layer 1
W1=tf.Variable(tf.random_normal([3,3,1,32],stddev=0.01))
L1=tf.nn.conv2d(X_img,W1,strides=[1,1,1,1],padding="SAME")
L1=tf.nn.relu(L1)
L1=tf.nn.max_pool(L1,ksize=[1,2,2,1],strides=[1,2,2,1],padding="SAME")
## 14*14*32


## Convolutional w. dropout Layer 2
W2=tf.Variable(tf.random_normal([3,3,32,64],stddev=0.01))
L2=tf.nn.conv2d(L1,W2,strides=[1,1,1,1],padding="SAME")
L2=tf.nn.relu(L2)
L2=tf.nn.max_pool(L2,ksize=[1,2,2,1],strides=[1,2,2,1],padding="SAME")
## 7*7*64

## Convolutional w. dropout Layer 3
W3=tf.Variable(tf.random_normal([3,3,64,128],stddev=0.01))
L3=tf.nn.conv2d(L2,W3,strides=[1,1,1,1],padding="SAME")
L3=tf.nn.relu(L3)
## 7*7*128

## Dense Layer 
L3=tf.reshape(L3,[-1,7*7*128])
W4=tf.Variable(tf.random_normal([7*7*128,10],stddev=0.01))
B4=tf.Variable(tf.random_normal([10]))

### Preidctin the result before SoftMax
predicted=tf.nn.softmax(tf.matmul(L3,W4)+B4)

## Weight initializer
##initializer=tf.contrib.layers.xavier_initializer()
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=predicted, labels=Y))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)

init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

def getBatch(i, size):
    startIndex = (i * size) % 59000
    endIndex = startIndex + size
    batch_X = x_train[startIndex : endIndex]
    batch_Y = y_train[startIndex : endIndex]
    return batch_X, batch_Y

def training_step(i):

    size = 128
    batch_X, batch_Y = getBatch(i, size)

    _,l=sess.run([optimizer,cost], feed_dict={X: batch_X, Y: batch_Y})
    print("Loss: %f" %l)

# Run number of iterations training the NN    
for i in range(50): 
    training_step(i)

writer = tf.summary.FileWriter('./graph', sess.graph)
pred=tf.argmax(predicted, 1)
preds=sess.run(pred,feed_dict={X:x_test})

```
